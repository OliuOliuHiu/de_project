```python
import pandas as pd
import os
import sys
sys.path.append(os.path.abspath("E:\de_work\project\mini\etl_basic").replace("\\", "/"))
from src.extract import fetch_users, get_data_csv
```

```python
df_api = fetch_users()
df_csv = pd.read_csv('../../data/orders.csv')
```

```python
df_api.columns.tolist()
```

```python
df_api = df_api.rename(columns = {"id": "user_id" , "name" : "customer_name"})
```

```python
df_api['user_id'] = 'uc' + df_api['user_id'].astype(str)
```

```python
df_api['city'] = df_api.apply(lambda row: row['address']['city'], axis=1)
df_api['company_name'] = df_api.apply(lambda row: row['company']['name'], axis=1)
```

```python
df_api_main = df_api[['user_id', 'customer_name','email','phone', 'city', 'company_name']]
```

```python
df_csv['user_id'] = 'uc' + df_csv['user_id'].astype(str)
```

```python
df_api_main
```

```python
df_csv
```

```python
df_fact_sales = pd.merge(df_csv, df_api_main, left_on= 'user_id', right_on= 'user_id', how = 'left')
```

```python
df_fact_sales
```

```python
df_fact_sales = df_fact_sales.fillna('Unknown')
```

```python
df_fact_sales['total_amount'] = df_fact_sales['quantity'] * df_fact_sales['price']
```

```python
df_fact_sales.columns.to_list()
```

```python
df_fact_sales['order_date'] = pd.to_datetime(df_fact_sales['order_date'])
df_fact_sales['order_month'] = df_fact_sales['order_date'].dt.strftime("%Y-%m")
```

```python
df_fact_sales
```

```python
columns_list = ['order_id','user_id','customer_name','city', 'product', 'quantity','total_amount', 'order_month']
df_fact_sales = df_fact_sales[columns_list]
```
